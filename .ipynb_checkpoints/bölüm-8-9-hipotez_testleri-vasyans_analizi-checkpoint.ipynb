## İş Uygulaması : Dönüşüm Oranı Testi

## Problem: 
* bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oran elde edildiği ifade edilmiş. Fakata bu durum kontrol edilmek isteniyor.Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.

## Detaylar:
* 500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alışveriş yapmış.
* örnek üzerinden elde edilen dönüşüm oranı: 40/500 = 0,08


* H0: p = 0.125 

* H1: p != 0.125

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40
nobs = 500
value = 0.125
```

```python
proportions_ztest(count, nobs, value)
# 1.argüman: başarı sayısı, 2.argüman: gözlem sayısı, 3. ise test edilmek istenen değeri)
```

* pvalue değeri 0.05 ten küçük H0 red.
* ürün satın alımına yönelik olarak test ettiğimiz 0.125 değerinin böyle olmadığı ,
* bu değerin yanlış olduğu bilgisine erişmiş oluyoruz. %95 güven ile söylemiş oluyoruz.

## Bağımsız İki Örneklem T Testi

* iki grup ortalaması arasında karşılaştırrma yapılma istenildiğinde kullanılır.

## Varsayımlar:
  * Normallik
  * Varyans Homojenliği

## PROBLEM:
   * Bir ML projesine yatırım yapılmış. Ürettiği tahminler neticesinde oluşan gelir ile eski sistemin ürettiği gelirler karşılaştırılıp anlamlı farklılık olup olmadığı test edilmek isteniyor.
## Detaylar:
* Model geliştirilmiş ve web sitesine entegre edilmiş.
* Site kullanıcıları belirli bir kurala göre 2 ye bölünmüş olsun.
* A grubu eski B grubu yeni sistem.
* Gelir anlamında anlamlı bir iş yapılıp yapılmadığını test edilmek isteniyor.

* H0: M1 = M2 

* H1: M1 != M2

```python
# VERI TIPI I
```

```python
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])


A_B = pd.concat([A, B], axis = 1)
A_B.columns = ["A","B"]

A_B.head()
```

```python
# VERI TIPI II
```

```python
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,
        25,21,18,24,26,33,26,27,28,19,25])

B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,
        32,33,33,33,31,32,33,26,32,33,29])

# A ve A'nın grubu
GRUP_A = np.arange(len(A))
GRUP_A = pd.DataFrame(GRUP_A)
GRUP_A[:] = "A"
A = pd.concat([A, GRUP_A], axis = 1)

# B ve B'nin Grubu
GRUP_B = np.arange(len(B))
GRUP_B = pd.DataFrame(GRUP_B)
GRUP_B[:] = "B"
B = pd.concat([B, GRUP_B], axis = 1)

# Tum veri
AB = pd.concat([A,B])
AB.columns = ["gelir","GRUP"]
print(AB.head())
print(AB.tail())
```

```python
import seaborn as sns
sns.boxplot(x = "GRUP", y = "gelir", data = AB);
```

## normallik varsayımı

```python
shapiro(A_B.A) # varsayımı sağlıyor. pvalue küçük değil. H0 reddedilemez
```

## varyans homojenligi varsayımı

* H0: Varyanslar Homojendir

* H1: Varyanslar Homojen Değildir

```python
stats.levene(A_B.A, A_B.B) 
# H0 reddedilemez. hesapladığımız pvalue değer alfa değerinden küçük olmadığı için H0 hipotezini Redddedemiyoruz.
# Varyanslar homojendir.
```

## Hipotez Testi

```python
stats.ttest_ind(A_B["A"], A_B["B"], equal_var = True) 
```

* varyans homojenliği varsayımının  sağlanıp sağlanmadığı true ile kontro ediliyor.

* pvalue değeri 0.05 ten küçük olduğundan dolayı farklılık vardır ve 
* farklılık olmadığını idda eden H0 hipotezi reddedilir.
* yani ana kitle ortalamaları birbirine eşitir diyen H0 hipotezi reddeddilir.

```python
test_istatistigi, pvalue = stats.ttest_ind(A_B["A"], A_B["B"], equal_var=True)
print('Test İstatistiği = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```

* Eski sistem ile yeni sistem arasında gelirler geö önünde bulundurulduğunda istatistiki olarak anlamlı bir farklılık vardır.
* Bu fark yeni sistemin lehinedir. Çünkü ortalamalarını incelediğimizde B grubun yani yeni grubun makine öğrenmesine göre çalışan grubun ortalaması daha yukardaydı .

## Nonparametrik Bağımsız İki Örneklem Testi
* hem normallik varsayımı hemde varyans homojenliği varsayımı sağlanmıyorsa bunu kullanırız.

```python
stats.mannwhitneyu(A_B["A"], A_B["B"])
```

```python
test_istatistigi, pvalue = stats.mannwhitneyu(A_B["A"], A_B["B"])
print('Test İstatistiği = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```

## Detaylar:

```python
oncesi = pd.DataFrame([123,119,119,116,123,123,121,120,117,118,121,121,123,119,
            121,118,124,121,125,115,115,119,118,121,117,117,120,120,
            121,117,118,117,123,118,124,121,115,118,125,115])

sonrasi = pd.DataFrame([118,127,122,132,129,123,129,132,128,130,128,138,140,130,
             134,134,124,140,134,129,129,138,134,124,122,126,133,127,
             130,130,130,132,117,130,125,129,133,120,127,123])
```

```python
oncesi[0:5]
```

```python
sonrasi[0:5]
```

```python
np.arange(len(oncesi))
```

```python
# BIRINCI VERI SETI
AYRIK = pd.concat([oncesi, sonrasi], axis = 1)
AYRIK.columns = ["ONCESI","SONRASI"]
print("'AYRIK' Veri Seti: \n\n ", AYRIK.head(), "\n\n")
```

```python
# IKINCI VERI SETI
# ONCESI FLAG/TAG'INI OLUSTURMA
GRUP_ONCESI = np.arange(len(oncesi))
GRUP_ONCESI = pd.DataFrame(GRUP_ONCESI)
GRUP_ONCESI[:] = "ONCESI"

# FLAG VE ONCESI DEGERLERINI BIR ARAYA GETIRME
A = pd.concat([oncesi, GRUP_ONCESI], axis = 1)

# SONRASI FLAG/TAG'INI OLUSTURMA
GRUP_SONRASI = np.arange(len(sonrasi))
GRUP_SONRASI = pd.DataFrame(GRUP_SONRASI)
GRUP_SONRASI[:] = "SONRASI"

# FLAG VE SONRASI DEGERLERINI BIR ARAYA GETIRME
B = pd.concat([sonrasi, GRUP_SONRASI], axis = 1)

# TUM VERIYI BIR ARAYA GETIRME
BIRLIKTE = pd.concat([A,B])
BIRLIKTE

# ISIMLENDIRME
BIRLIKTE.columns = ["PERFORMANS","ONCESI_SONRASI"]
print("'BIRLIKTE' Veri Seti: \n\n", BIRLIKTE.head(), "\n")
```

```python
A = pd.DataFrame([28,33,30,29,28,29,27,31,30,32,28,33,25,29,27,31,31,30,31,34,30,32,31,34,28,32,31,28,33,29])

B = pd.DataFrame([31,32,30,30,33,32,34,27,36,30,31,30,38,29,30,34,34,31,35,35,33,30,28,29,26,37,31,28,34,33])

C = pd.DataFrame([40,33,38,41,42,43,38,35,39,39,36,34,35,40,38,36,39,36,33,35,38,35,40,40,39,38,38,43,40,42])

dfs = [A, B, C]

ABC = pd.concat(dfs, axis = 1)
ABC.columns = ["GRUP_A","GRUP_B","GRUP_C"]
ABC.head()
```

## Varsayım Kontrolü

```python
from scipy.stats import shapiro
# normallik varsayımı için
```

* H0: örnek dağılımı ile teorik normal dağılım arasında istatiski olarak anlamlı bir fakrlılık yoktur.
* 3 grup içinde normallik varsayımı sağlanıyor çünkü pvalue 0.05 ten büyük.H0 reddedilemez.

```python
shapiro(ABC["GRUP_A"])
```

```python
shapiro(ABC["GRUP_B"])
```

```python
shapiro(ABC["GRUP_C"])
```

```python
stats.levene(ABC["GRUP_A"], ABC["GRUP_B"],ABC["GRUP_C"]) 
#homojenlik varsayımı
#varyanslar homojendir. pavlye 0.05 ten büyük ve H0 reddedilemez.
```

## Hipotez Testi

```python
from scipy.stats import f_oneway
```

```python
f_oneway(ABC["GRUP_A"], ABC["GRUP_B"],ABC["GRUP_C"])
```

```python
print('{:.5f}'.format(f_oneway(ABC["GRUP_A"], ABC["GRUP_B"],ABC["GRUP_C"])[1]))
```

* H0 red. gruplar arasında istatiski olarak anlamlı bir farklılık vardır.
* farklılığını bildiğimiz grubun lehine olacak şekilde seçme işlemini gerçekleştirebiliriz.

```python
ABC.describe().T
```

* C grubunun diğer grublara göre daha anlamlı katkılar sağlayabileceği gözlemlenmektedir. 

## Nonparametrik Hipotez Testi

```python
from scipy.stats import kruskal
```

```python
kruskal(ABC["GRUP_A"], ABC["GRUP_B"],ABC["GRUP_C"])
```

* H0 reddedilir.gruplar arasında en az birisi diğerlerinden farkldır.
