## Veri Okuryazarlığı Nedir?

* Her türden veri tipini, değişken ve ölçek türlerini tanımlayabilme, betimsel istatistikleri ve istatistiksel grafikleri kullanarak veri değerlendirebilme yeteneğidir.

* Verinin tamamına popülasyon, bu veriyi temsil eden alt kümeye örneklem denir. Buradaki örneklemin bu ana kitleyi yani popülasyonu çok iyi temsil etmesi gerekiyor.
* elde ettiğimiz örneklem popülasyonun niteliklerini, özelliklerini temsil etmesi ve yansız olması gerekir.

## Değişken Türleri
* ***Sayısal Değişkenler(nicel, kantitatif)***
* ***Kategorik Değişkenler(nitel, kalitatif)***

## Ölçek Türleri
* ***Sayısal değişkenler için: Aralık ve Oran***
* ***Kategorik değişkenler için: Nominal ve Ordinal***

* ***sayısal değişkenler:***
* 1. ***ARALIK:*** Başlangıç noktası sıfır olmayan sayısal değişkenlerin ölçek türüdür.
* 2. ***Oran:*** Başlangıç noktasını sıfır kabul eden sayısal değişkenlerin ölçek türüdür.

* ***Kategorik değişkenler:***
* 1.***Nominal:*** kategorik değişkenlerin sınıfları arasında fark olmayınca
* 2.***Ordinal:*** kategorik değişkenlerin sınıfları arasında fark var olunca

## Aritmetik Ortalama
* bir seride (değişkende) yer alan tüm değerlerin toplanması ve birim sayısına bölünmesi ile elde edilen istatistiktir.

## Medyan
* Bir seriyi küçükten büyüğe veya büyükten küçüğe sıraladığımızda tam orta noktadan seriyi iki eşit parçaya ayıran değere denir.

* Aritmetik ortalama, seri dağılımının (değişkenin dağılımının) simektrik olduğu bilindiğinde kullanılabilir.
* simetrik değilse bu durumda medyan hesaplanabilir.

## Mod
* Bir seride (değişkende) en çok tekrar eden değere denir.

## Kartiller (Quartiles)
* Küçükten büyüğe sıralanan bir seriyi 4 parçaya ayıran değerlere denir.

* 1. çeyrek değeri:10, 2. çeyrek değeri: 14, 3. çeyrek değeri: 17

* Serinin merkezi eğilimiyle ve dağılımıyla ilgili bilgi almak için kullanılır.

## Merkezi eğilimin önemini kavrama

* 1. Temsil yönünü kavrama
* 2. Doğru kullanılması

* portföy temsil istatistiklerine bakılır.
* A'nın ortalaması: 706000, B'nin ortalaması: 730000

* A'nın medyanı: 700000, B'nin medyanı: 400000
* B de medyanı kullanmak daha iyi olur.
